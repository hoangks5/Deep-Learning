```python
# Bài 3
# Gradient descent là thuật toán tìm giá trị nhỏ nhất của hàm số f(x) dựa trên đạo hàm . Thuật toán:
# Khởi tạo giá trị x = x0 tùy ý
# Gán x = x - rate_learning*f'(x)
# Tính lại f(X) và lặp lại cho đến khi giá trị f(x) đử nhỏ thì dừng lại , ngược lại thì làm tiếp bước 2
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = 20
y = x*x +2*x+5
```

```python
rate_learning = 0.1
y = []
for i in range(20):
  x = x - (2*x +2)*rate_learning
  y.append(2*x+2)
plt.plot(y)
plt.xlabel('So lan')
plt.ylabel('Fx')
plt.show()
```

```python
# Bài 4
# X = [x1,x2,x3,...xm].T
# x = [1,x1,x2,...xm]
# w = [w0,w1,w2,...wm].T
# x*w = x*w = w0 + x1*w1 + x2*w2 +...+ xm*wm
# y = [y1,y2,y3,...ym]
# Khi đó w = (X.T*X)^-1*X.T*y
# Khi X.T*X khả nghịch
# w có dạng vector w = [w0,w1,w2,...wm]
# Suy ra được các hệ số của hàm xấp xỉ f(X)
```

```python
# Bài 5
def plot(x):
  rate_learning = x
  y = []
  for i in range(20):
    x = x - (2*x +2)*rate_learning
    y.append(2*x+2)
  plt.plot(y)
  plt.xlabel('So lan')
  plt.ylabel('Fx')
  plt.show()
plot(0.2)
plot(0.1)
plot(0.01)
```
