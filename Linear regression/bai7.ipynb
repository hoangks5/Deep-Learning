```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/nttuan8/DL_Tutorial/master/L1/data_square.csv').values
```

```python
N = data.shape[0] # Đếm số lượng dữ liệu
# N = 30
x = data[:,0].reshape(-1,1) # Lấy dữ liệu x với x là diện tích
y = data[:,1].reshape(-1,1) # Lấy dữ liệu y là giá nhà
x = np.hstack((np.ones((N,1)), x)) # Tạo vector dạng [1,x]
w = np.array([0.,1.]).reshape(-1,1) # Khởi tạo vector w với w0 = 0, w1 = 1
num = 1000
cost = np.zeros((num,1)) # Tạo vector 1 1x100 [0]
rate_learning = 0.000001 # Giá trị rate 
for i in range(1,num):
  r = np.dot(x, w) - y # Giá trị (y -yci)
  cost[i] = 0.5*np.sum(r*r) # Gán giá trị tính tổng loss function = E(0.5(y-yci)^2)
  w[0] = w[0] - rate_learning*np.sum(r) 
  w[1] = w[1] - rate_learning*np.sum(np.multiply(r,x[:,1].reshape(-1,1)))
  print(cost[i])
```

```python
y = w[0] + 50*w[1] # Thử tính với diện tích là 50
print(y)
```
